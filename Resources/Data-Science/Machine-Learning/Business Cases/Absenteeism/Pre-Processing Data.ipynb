```python
df.drop(columns='ID',inplace=True)
```

```python
dfnew = pd.get_dummies(df,columns = ['Reason for Absence'])
```

```python
dfnew.columns
```

```python
Reason_1 = ['Reason for Absence_1','Reason for Absence_2','Reason for Absence_3','Reason for Absence_4','Reason for Absence_5','Reason for Absence_6','Reason for Absence_7','Reason for Absence_8','Reason for Absence_9','Reason for Absence_10','Reason for Absence_11','Reason for Absence_12','Reason for Absence_13','Reason for Absence_14']
Reason_2 =['Reason for Absence_15','Reason for Absence_16','Reason for Absence_17']
Reason_3 = ['Reason for Absence_18','Reason for Absence_19','Reason for Absence_21']
Reason_4 = ['Reason for Absence_22','Reason for Absence_23','Reason for Absence_24','Reason for Absence_25','Reason for Absence_26','Reason for Absence_27','Reason for Absence_28']
```

```python
dfnew.drop(columns='Reason for Absence_0',inplace=True)
```

```python
Group_1 = dfnew[Reason_1].max(axis=1)
Group_2 = dfnew[Reason_2].max(axis=1)
Group_3 = dfnew[Reason_3].max(axis=1)
Group_4 = dfnew[Reason_4].max(axis=1)
```

```python
groups = Reason_1 + Reason_2+Reason_3+Reason_4
groups
dfnew.drop(columns=groups,inplace=True)
```

```python
dfnew['Reason_1'] = Group_1
dfnew['Reason_2'] = Group_2
dfnew['Reason_3'] = Group_3
dfnew['Reason_4'] = Group_4
```
