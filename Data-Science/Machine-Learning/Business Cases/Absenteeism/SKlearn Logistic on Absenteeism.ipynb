```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
sns.set()
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Absenteeism-preprocessed.csv')
```

```python
targets = np.where(df['Absenteeism Time in Hours']>df['Absenteeism Time in Hours'].median(),1,0)
df
```

```python
df.drop(columns = 'Absenteeism Time in Hours',inplace = True)
```

```python
df
```

```python
targets.sum()/targets.shape[0]

xcols =df.iloc[:,:-1] #all columns but last 
y=targets
cols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pet']
icols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Day of the Week', 'Transportation Expense', 'Age',
       'Body Mass Index', 'Education', 'Children',
       'Pet']
allcols=cols+['Absenteeism Time in Hours']
x1 = df[cols]
```

```python
cols2scale = ['Day of the Week', 'Transportation Expense', 'Age',
       'Body Mass Index', 'Children',
       'Pet']
nonscale = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education']
x1 = df[cols2scale]
```

```python
scaler = StandardScaler()
scaled_x1 = scaler.fit_transform(x1)
scaled_x1 = pd.DataFrame(scaled_x1, columns=cols2scale)
```

```python
ns =df[nonscale]
```

```python
x = pd.concat([scaled_x1,ns],axis=1)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8)
```

```python
clf = LogisticRegression().fit(x_train,y_train)
clf.score(x_train,y_train)
```

```python
clf.intercept_
```

```python
clf.coef_
```

```python
allcols = x.columns.values
```

```python
sumdf = pd.DataFrame(columns=['Features'],data = allcols)
sumdf['Coef'] = np.transpose(clf.coef_)
sumdf.index = sumdf.index +1
sumdf.loc[0] = ['Intercept', clf.intercept_[0]]
sumdf.sort_index(inplace=True)
sumdf
```

```python
sumdf['odds']= np.exp(sumdf['Coef'])
```

```python
sumdf.sort_values('Coef',axis=0,ascending=False,inplace=True)
```

```python
sumdf
```

```python
clf.score(x_test,y_test)
```
